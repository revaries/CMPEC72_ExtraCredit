```python
# This is to check which city has most payments claimed
city = {}
for each in InMatrix[1:]:
    cityval = each[4]
    amount = each[10]
    if cityval not in city:
        city[cityval] = float(amount[1:])
    else:
        city[cityval] = city[cityval]+float(amount[1:])
import operator
mostamountclaimed = max(city.iteritems(), key=operator.itemgetter(1))[0]
amount = city[max(city.iteritems(), key=operator.itemgetter(1))[0]]
print "%s is the city with max payments claimed. And it has claimed a total of %s"%(mostamountclaimed,amount)
```

```python
# Converting Key value pairs to array
x = [[None]*500]*2
for key,value in city.iteritems():
    temp = [key,value]
    x.append(temp)
```

```python
#Printing Whole Output
FinalArray = [[None]*500]*2 
FinalArray = x[2:]
print FinalArray
```

```python
FinalArrayCityNames =[None]*0
FinalArrayClaimed =[None]*0
for each in FinalArray:
    FinalArrayCityNames.append(each[0])
    FinalArrayClaimed.append(each[1])
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(len(FinalArrayClaimed))
plt.bar(x, FinalArrayClaimed)
plt.xticks(x + 0.5,FinalArrayCityNames, rotation=90)


```
